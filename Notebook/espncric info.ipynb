```python
total_players=[]
def get_total_match_players(link):
    print(link)
    agent = {"User-Agent":'Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36'}
    r = requests.get(link, headers=agent)
    soup = BeautifulSoup(r.content, 'html5lib')
    table = soup.find_all('a')
    if len(table) != 0:
        total_players = []
        for i in table:
            if type(i.get('title')) == str:
                total_players.append(i.get('title'))
        return total_players
    else:
        return None
```

```python
def teams_playing_XI(total_players):
    first_innings_batters = []
    first_innings_bowlers = []
    second_innings_batters = []
    second_innings_bowlers = []
    first_innings_toggle = True
    second_innings_toggle = False

    if len(total_players) != 0:
        for i in total_players:
            if "View full profile of" in i and second_innings_toggle == False:
                first_innings_toggle = False
                first_innings_bowlers.append(i.lstrip("View full profile of"))
            elif first_innings_toggle == False and "View full profile of" not in i:
                second_innings_toggle = True
                second_innings_batters.append(i)
            elif first_innings_toggle == True:
                first_innings_batters.append(i)
            elif first_innings_toggle == False and second_innings_toggle == True and "View full profile of" in i:
                second_innings_bowlers.append(i.lstrip("View full profile of"))
        
        #Combing team based players
        result_list_Team1 = [first_innings_batters,second_innings_bowlers]
        result_list_Team2 = [second_innings_batters,first_innings_bowlers]
        
        set_Team_1 = set().union(*result_list_Team1)
        set_Team_2 = set().union(*result_list_Team2)
        
        return set_Team_1,set_Team_2
    else:
        return None
```

```python
lst = a_2022['Total_Players'].apply(teams_playing_XI)
```

```python
Team1_list = []
Team2_list = []
for i in range(len(lst)):
    if lst[i] != None:
        Team1_list.append(lst[i][0])
        Team2_list.append(lst[i][1])
    else:
        Team1_list.append(None)
        Team2_list.append(None)
```

```python
import re
p = ":_?([0-9]+)"
def get_target_details(link):
    print(link)
    agent = {"User-Agent":'Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36'}
    r = requests.get(link, headers=agent)    # Reading the HTML Page
    #print(r.content)
    
    soup = BeautifulSoup(r.content, 'html5lib') # Loading the HTMl to a beautifulSoup object
    #print(soup.prettify())
    
    table = soup.find_all('span',attrs={"class":"ds-text-tight-s ds-font-regular"},text=True)
    if len(table) != 0:
        s = table[0].text
        a = re.findall(p,s)
        return a
    else:
        return None
```
