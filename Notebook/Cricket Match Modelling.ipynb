```python
def toss_decision(toss):
    if ',' not in toss:
        return None
    else:
        a,b = toss.split(',')
        if "bat first" in toss:
            return a + " bat first"
        elif "field first" in toss:
            return a + " field first"
```

```python
def bat_first(toss_decision,Team1,Team2):
    print(toss_decision,Team1,Team2)
    if toss_decision != None:
        if Team1 in toss_decision and 'bat first' in toss_decision:
            return Team1
        elif Team1 in toss_decision and 'field first' in toss_decision:
            return Team2
        elif Team2 in toss_decision and 'bat first' in toss_decision:
            return Team2
        elif Team2 in toss_decision and 'field first' in toss_decision:
            return Team1
    else:
        return None
```

```python
import re,html
def clean(text):
    #tags like <tag>
    text = re.sub(r'<[^<>]*>', ' ',text)
    #Markdown Urls
    text = re.sub(r'\[([^\[\]]*)\]\([^\(\)]*\)',r'\1',text)
    #Remove Punctuation
    text = re.sub(r'([!?,])\1+', r'\1', text)
    #Remove all URL's
    text = re.sub(r'http.*', ' ', text)
    #Remove @
    text = re.sub(r'@\w*', ' ', text)
    #text or code in brackets
    text = re.sub(r'\[[^\[\]]*\]',' ',text)
    # remove b"
    text = text.replace('b\"',' ') 
    # remove b'
    text = text.replace("\'",' ') 
    # remove \\n
    text = text.replace('{',' ')
    text = text.replace('}',' ')
    #Remove &amp
    text = text.replace('&amp',' ') 
    # remove UTF-8 code like \\xe2
    text = re.sub(r'(\\x(.){2})', ' ',text) 
    #Standalone sequences for specials
    text = re.sub(r'(?:^|\s)[;.\'\"&#<>{}\[\]+|\\:-]{1,}(?:\s|$)', ' ',text)
    #stand alone sequence of hyphens 
    text= re.sub(r'(?:^|\s)[\-=\+]{2,}(?:\s|$)', ' ',text)
    # Sequence of white spaces
    text = re.sub(r'\s+',' ',text)
    return text.strip('"')
```

```python
def cleaned_total_players(tp):
    lst = []
    a = clean(tp)
    for i in a.split(','):
        i = i.lstrip()
        i = i.rstrip()
        i = i.strip('"')
        lst.append(i)
    return lst
```

```python

wc_2022_playing_teams = ['South Africa', 'Bangladesh', 'Pakistan','England','Afghanistan', 'Australia',
            'New Zealand','India','U.A.E.','Namibia','Netherlands','Sri Lanka','Ireland','Scotland','West Indies','Zimbabwe']
team_ids = {'India': 16,'England': 15,'Pakistan':14, 'South Africa':13,'New Zealand':12,'Australia':11,'West Indies':10,
           'Sri Lanka':9,'Bangladesh':8,'Afghanistan':7,'Zimbabwe':6,'U.A.E':5,'Ireland':4,'Namibia':3,'Scotland':2,
            'Netherlands':1}

def get_match_prediction(Team1,Team2,Toss,Target,bat_first,Ground,total_players):
    if len(total_players) > 22:
        return "Enterred Invalid Number of Players"
    elif Team1 and Team2 not in wc_2022_playing_teams:
        return "Entered Invalid Teams"
    elif Toss not in (Team1,Team2):
        return "Entered Invalid Toss Team"
    elif bat_first not in (Team1,Team2):
        return "Entered Invalid bat_first Team"
    elif Target <=50 | Target>250 :
        return "Enter a Valid Target"
    
    team1_id = team_ids.get(Team1)
    team2_id = team_ids.get(Team2)
    bat_first_id = team_ids.get(bat_first)
    toss_id = team_ids.get(Toss)
    
    d['Team 1'] = team1_id
    d['Team 2'] = team2_id
    d['bat_first'] = bat_first_id
    d['Toss'] = toss_id
    d['Target'] = Target
    
    grd = "Ground_" + Ground
    
    for i in Columns_list:
        if i == grd:
            d[i] = 1
        
        for j in total_players:
             if i == j:
                    d[i] = 1
            
    df = pd.DataFrame(data=d.values())
    test_pca = pca.transform(df.T)
    win_id = best_grid.predict_proba(test_pca)
    return win_id
    #if win_id[0] == 0:
    #    return Team1
    #elif win_id[0] == 1:
    #    return Team2
```
